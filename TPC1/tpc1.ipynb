* Crie uma função que calcula a distribuição da doença por níveis de colesterol. Considere um nível igual a um intervalo de 10 unidades, comece no limite inferior e crie os níveis necessários até abranger o limite superior;

```python
def carregarBD(fnome):
    db = []
    file_csv = open(fnome)
    
    newLine1 = file_csv.readline()
    cabecalhoAux = newLine1.replace("\n","")
    cabecalho = cabecalhoAux.split(",")

    for line in file_csv:
        
        myDic = {}
        newLine = line.replace("\n","")
        campos = newLine.split(",")
        
        for i in range(0,len(cabecalho)):
            myDic[cabecalho[i]] = campos[i]
        db.append(myDic)
    return db
```

**Distribuição de doenças por sexo**

```python
def dist_Doenca_Sexo(db):
    
    doentes_Masculino = 0
    doentes_Feminino = 0
    nDoentes_Masculino = 0
    nDoentes_Feminino = 0
    
    result_doentes = {}
    result_NDoentes = {}
        
    for dict in db:
                            
        if dict['sexo'] == 'M':
            if dict['temDoença'] == '1':
                doentes_Masculino += 1
            else:
                nDoentes_Masculino += 1
                    
        if dict['sexo'] == 'F': 
            if dict['temDoença'] == '1':
                doentes_Feminino += 1 
            else:
                nDoentes_Feminino +=1
    
    result_doentes['F'] = doentes_Feminino
    result_doentes['M'] = doentes_Masculino
    result_NDoentes['F'] = nDoentes_Feminino
    result_NDoentes['M'] = nDoentes_Masculino
    
    return (result_doentes,result_NDoentes)
```

**Distribuição de doenças por escalão étario**

```python
maxIdade = 0

def dist_Doenca_Idade(db):
    for dict in db:
        global maxIdade
        if maxIdade < int(dict['idade']) and dict['temDoença'] == '1':
            maxIdade = int(dict['idade'])
    
    idade_doentes = {}
    idade_NDoentes = {}
    
    for dict in db:
        if dict['temDoença'] == '1':
            if dict['idade'] not in idade_doentes:
                idade_doentes[dict['idade']] = 1
        
            else:
                idade_doentes[dict['idade']] += 1
        else:
            if dict['idade'] not in idade_NDoentes:
                idade_NDoentes[dict['idade']] = 1
        
            else:
                idade_NDoentes[dict['idade']] += 1

    return (idade_doentes,idade_NDoentes)
```

**Distribuição de doenças por níveis de colesterol**

```python
limite_inferior = 0
limite_superior = 0

def dist_Doenca_Colesterol(db):
    
    global limite_inferior
    limite_inferior = int(db[0]['colesterol'])
    global limite_superior
    limite_superior = int(db[0]['colesterol'])
        
    colesterol_doentes = {}
    colesterol_NDoentes = {}
     
    for dict in db:
        if dict['temDoença'] == '1':
            if limite_superior < int(dict['colesterol']):
                limite_superior = int(dict['colesterol'])
            if int(dict['colesterol']) < limite_inferior:
                limite_inferior = int(dict['colesterol'])
            
            
            if dict['colesterol'] not in colesterol_doentes:
                colesterol_doentes[dict['colesterol']] = 1
        
            else:
                colesterol_doentes[dict['colesterol']] += 1
        else:
            if dict['colesterol'] not in colesterol_NDoentes:
                colesterol_NDoentes[dict['colesterol']] = 1
        
            else:
                colesterol_NDoentes[dict['colesterol']] += 1
             
    return (colesterol_doentes,colesterol_NDoentes)
```

**Tabelas de Distribuição**

```python
def tabela_distribuicao(state,tuplo):
    
    data = []
    primeira_linha = [' ','Doente','Não Doente']
    data.append(primeira_linha)
    
    if state == 1:
        for key in tuplo[0]:
            if key == 'F':
                row = ['Feminino',tuplo[0][key],tuplo[1]['F']]
                data.append(row)
                
            if key == 'M':
                row = ['Masculino',tuplo[0][key],tuplo[1]['M']]
                data.append(row)
                
    
    if state == 2:
        first = 30
        last = maxIdade
        
        age_range_doentes = {}
        age_range_NDoentes = {}

        temp = first
        for i in range(int((last-first)/5)):
            aux = temp + 4
            age_range_doentes[f'{temp}-{aux}'] = 0
            temp += 5
                    
        if temp != last:
            age_range_doentes[f'{temp}-{last}'] = 0    
               
        for key in tuplo[0]:
            if int(key) % 5 != 0:
                ind_menor = int(key) - (int(key) % 5)
                ind_maior = ind_menor + 4
                if f'{ind_menor}-{ind_maior}' in age_range_doentes.keys():
                    age_range_doentes[f'{ind_menor}-{ind_maior}'] += tuplo[0][key]
                else:
                    age_range_doentes[f'{last-(last % 5)}-{last}'] += tuplo[0][key]
            
            else:
                if f'{key}-{int(key)+4}' in age_range_doentes.keys():
                    age_range_doentes[f'{key}-{int(key)+4}'] += tuplo[0][key]
                else:
                    age_range_doentes[f'{last-(last % 5)}-{last}'] += tuplo[0][key]
                
        temp = first
        for i in range(int((last-first)/5)):
            aux = temp + 4
            age_range_NDoentes[f'{temp}-{aux}'] = 0
            temp += 5
                    
        if temp != last:
            age_range_NDoentes[f'{temp}-{last}'] = 0           
        
        for key in tuplo[1]:
            if int(key) % 5 != 0:
                ind_menor = int(key) - (int(key) % 5)
                ind_maior = ind_menor + 4
                if f'{ind_menor}-{ind_maior}' in age_range_NDoentes.keys():
                    age_range_NDoentes[f'{ind_menor}-{ind_maior}'] += tuplo[1][key]
                else:
                    age_range_NDoentes[f'{last-(last % 5)}-{last}'] += tuplo[1][key]
            
            else:
                if f'{key}-{int(key)+4}' in age_range_NDoentes.keys():
                    age_range_NDoentes[f'{key}-{int(key)+4}'] += tuplo[1][key]
                else:
                    age_range_NDoentes[f'{last-(last % 5)}-{last}'] += tuplo[1][key]
                
        for key in age_range_doentes:
            row = [f'{key}',age_range_doentes[key], age_range_NDoentes[key]]
            data.append(row)
    
    if state == 3:
        colesterol_range_doentes = {}
        colesterol_range_NDoentes = {}
        temp = limite_inferior
        for i in range(int((limite_superior-limite_inferior)/10)):
            aux = temp + 9
            colesterol_range_doentes[f'{temp}-{aux}'] = 0
            temp += 10
                    
        if temp != limite_superior:
            colesterol_range_doentes[f'{temp}-{limite_superior}'] = 0           
        
        for key in tuplo[0]:
            if int(key) % 10 != 0:
                ind_menor = int(key) - (int(key) % 10)
                ind_maior = ind_menor + 9
                if f'{ind_menor}-{ind_maior}' in colesterol_range_doentes.keys():
                    colesterol_range_doentes[f'{ind_menor}-{ind_maior}'] += tuplo[0][key]
                else:
                    colesterol_range_doentes[f'{limite_superior-(limite_superior % 10)}-{limite_superior}'] += tuplo[0][key]
            
            else:
                if f'{key}-{int(key)+9}' in colesterol_range_doentes.keys():
                    colesterol_range_doentes[f'{key}-{int(key)+9}'] += tuplo[0][key]
                else:
                    colesterol_range_doentes[f'{limite_superior-(limite_superior % 10)}-{limite_superior}'] += tuplo[0][key]
         
        temp = limite_inferior
        for i in range(int((limite_superior-limite_inferior)/10)):
            aux = temp + 9
            colesterol_range_NDoentes[f'{temp}-{aux}'] = 0
            temp += 10
                    
        if temp != limite_superior:
            colesterol_range_NDoentes[f'{temp}-{limite_superior}'] = 0           
        
        for key in tuplo[1]:
            if int(key) % 10 != 0:
                ind_menor = int(key) - (int(key) % 10)
                ind_maior = ind_menor + 9
                if f'{ind_menor}-{ind_maior}' in colesterol_range_NDoentes.keys():
                    colesterol_range_NDoentes[f'{ind_menor}-{ind_maior}'] += tuplo[1][key]
                else:
                    colesterol_range_NDoentes[f'{limite_superior-(limite_superior % 10)}-{limite_superior}'] += tuplo[1][key]
            
            else:
                if f'{key}-{int(key)+9}' in colesterol_range_NDoentes.keys():
                    colesterol_range_NDoentes[f'{key}-{int(key)+9}'] += tuplo[1][key]
                else:
                    colesterol_range_NDoentes[f'{limite_superior-(limite_superior % 10)}-{limite_superior}'] += tuplo[1][key]
               
        for key in colesterol_range_doentes:
            row = [f'{key}',colesterol_range_doentes[key], colesterol_range_NDoentes[key]]
            data.append(row)
        
      
    col_width = [max(len(str(x)) for x in col) for col in zip(*data)]

    for row in data:
        print("  ".join(str(x).ljust(width) for x, width in zip(row, col_width)))
        

def main():
    data_set = carregarBD("myheart.csv")
    result = input("Indique a distribuição que pretende efetuar:\n1 - Doença por sexo\n2 - Doença por escalão étario\n3 - Doença por nível de colesterol")
    if result == '1':
        tabela_distribuicao(1,dist_Doenca_Sexo(data_set))
    elif result == '2':
        tabela_distribuicao(2,dist_Doenca_Idade(data_set))
    elif result == '3':
        tabela_distribuicao(3,dist_Doenca_Colesterol(data_set))
    else:
        print("Distribuição não válida!\n")
        
main()
```
